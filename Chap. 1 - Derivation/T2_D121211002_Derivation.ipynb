For function 
$f_2=t^3+t$

1. Plot the function
2. Plot the derivative of the function (exact solution)
3. Calculate the numerical derivative of the function with finite different method
4. Compare number 2 and 3

```python
import numpy as np
```

```python
h = 0.1
```

```python
x = np.arange(-5, 5, h)
```

```python
# compute function
y = x**3 + x
```

```python
x
```

```python
y
```

```python
import matplotlib.pyplot as plt
```

### Answer for question No.1

```python
#1. Plot the function
plt.figure(figsize=(6,3))
plt.plot(x,y,label = 'function f(x)')
plt.legend()
plt.show()
```

### Answer for question No.2

```python
#2. Plot the derivative of the function (exact solution)
plt.figure(figsize=(6,3))
plt.plot(x,3*x**2 +1,label = 'f\'(x) exact/analytic solution')
plt.legend()
plt.show()
```

### Answer for question No.3

```python
#3. Numerical derivative of the function with finite difference method
plt.figure(figsize=(6,3))
plt.plot(x[:-1:],np.diff(y)/h,label = 'f\'(x) numerical solution from finite difference')
plt.legend()
plt.show()
```

### Answer for question No.4

```python
#4.Compare 2 and 3
plt.figure(figsize=(6,3))
plt.plot(x,3*x**2 +1,label = 'f\'(x) exact/analytic solution')
plt.plot(x[:-1:],np.diff(y)/h,label = 'f\'(x) numerical solution from finite difference')
plt.legend()
plt.show()
```
